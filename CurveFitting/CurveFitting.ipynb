```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## make a datasets

```python
x = np.array([((i+1+ np.random.randn())/1000 ) for i in range(100000)])
y = 0.5*(x**2) + np.log(x) - 100*np.sin(4*x)
plt.plot(x,y)
```

## use scipy for curve fitting

```python
# curve fittingの関数を読み込み
from scipy.optimize import curve_fit

# fittingさせる近似式を用意
def func_fitting(x,a,b,c,d,e):
    return x + a*(x**3) + b*(x**5) + c*(x**7) + d*(x**9) + e*(x**11)
    
# curve fittingでは係数と係数の共分散を返す
eff, cov = curve_fit(func_fitting, x, y,method='lm',maxfev=100)

y_pred = func_fitting(x,eff[0],eff[1],eff[2],eff[3],eff[4])
```

```python
print('Scipy関数での二乗誤差(SE)：', np.sum((y_pred - y)**2))
print()
for i_eff in eff:
    print(i_eff)
print()
plt.plot(x,y,c='k')
plt.plot(x,y_pred,c='r')
plt.show()
```

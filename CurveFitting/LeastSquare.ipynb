## OLSによるcurvefittingの係数推定

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x,a):
    return a*x+3*(x**2+ x**3 + 3)

a_list = np.linspace(-10,-4,100)
x_list = np.linspace(-1,1,10000)
a_correct=-6.9501849817982362
y_list = np.array(func(x_list, a_correct))

se_list = []
for i_a in a_list:
    y_pred = np.array(func(x_list, a=i_a))
    SE = np.sum((y_list - y_pred)**2)
    se_list.append(SE)
se_list = np.array(se_list) 
```

```python
plt.scatter(a_list, np.array(se_list))
plt.show()
```

```python
np.where(se_list == np.min(se_list))
```
